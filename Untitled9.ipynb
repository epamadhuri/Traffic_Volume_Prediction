```python
# Step 1: Import pandas
import pandas as pd

# Step 2: Load the uploaded CSV into a DataFrame
df = pd.read_csv("Metro_Interstate_Traffic_Volume.csv")

# Step 3: Now you can explore the dataset
df.info()
df.describe()
df.isnull().sum()
```

```python
# Fill missing holidays with 'None'
df['holiday'] = df['holiday'].fillna('None')
```

```python
# Convert to datetime
df['date_time'] = pd.to_datetime(df['date_time'])

# Extract time-based features
df['hour'] = df['date_time'].dt.hour
df['day'] = df['date_time'].dt.day
df['month'] = df['date_time'].dt.month
df['day_of_week'] = df['date_time'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].isin([5,6]).astype(int)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df['date_time'], df['traffic_volume'])
plt.title("Traffic Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Traffic Volume")
plt.show()
```

```python
# --- Step 4: Advanced Feature Engineering ---

# Lag features (previous traffic volumes)
for lag in [1, 2, 3, 6, 12, 24]:
    df[f'lag_{lag}'] = df['traffic_volume'].shift(lag)

# Rolling mean & std (moving averages)
for window in [3, 6, 12, 24]:
    df[f'rolling_mean_{window}'] = df['traffic_volume'].shift(1).rolling(window=window).mean()
    df[f'rolling_std_{window}'] = df['traffic_volume'].shift(1).rolling(window=window).std()

# Cyclical encoding for hour and day of week
import numpy as np
df['hour_sin'] = np.sin(2 * np.pi * df['hour']/24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour']/24)
df['dow_sin'] = np.sin(2 * np.pi * df['day_of_week']/7)
df['dow_cos'] = np.cos(2 * np.pi * df['day_of_week']/7)

# Simple weather impact feature
df['weather_score'] = df['temp'] - (df['rain_1h']*5 + df['snow_1h']*10)

# Drop rows with NaN created by lag/rolling
df = df.dropna()

print("Feature engineering complete. Shape:", df.shape)
```

```python
# Drop non-numeric columns (or encode them separately)
X = df.drop(columns=['date_time','holiday','weather_main','weather_description','traffic_volume'])
y = df['traffic_volume']

# Ensure no NaN values remain
X = X.fillna(0)

# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
# --- Step 6: Evaluation & Optimization ---

import lightgbm as lgb
import xgboost as xgb
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define features and target again (safe reset)
X = df.drop(columns=['date_time','holiday','weather_main','weather_description','traffic_volume'])
y = df['traffic_volume']

# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Recreate and fit models
lgb_model = lgb.LGBMRegressor(n_estimators=500, learning_rate=0.05, subsample=0.8, colsample_bytree=0.8)
xgb_model = xgb.XGBRegressor(n_estimators=500, learning_rate=0.05, max_depth=8, subsample=0.8, colsample_bytree=0.8)

lgb_model.fit(X_train, y_train)
xgb_model.fit(X_train, y_train)

# Final blended predictions
final_pred = 0.55*lgb_model.predict(X_test) + 0.45*xgb_model.predict(X_test)

# Metrics
print("R² Score:", r2_score(y_test, final_pred))
print("MAE:", mean_absolute_error(y_test, final_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, final_pred)))

# Actual vs Predicted plot
plt.figure(figsize=(12,6))
plt.plot(y_test.values, label="Actual", alpha=0.7)
plt.plot(final_pred, label="Predicted", alpha=0.7)
plt.legend()
plt.title("Actual vs Predicted Traffic Volume")
plt.show()

# Residual distribution
residuals = y_test - final_pred
sns.histplot(residuals, bins=50, kde=True)
plt.title("Residual Distribution")
plt.show()

# Feature importance (LightGBM example)
lgb.plot_importance(lgb_model, max_num_features=20, importance_type='gain')
plt.title("LightGBM Feature Importance")
plt.show()
```

```python
import pandas as pd
import gradio as gr

# exact feature list from training
features = ['temp','rain_1h','snow_1h','clouds_all','hour','day','month',
            'day_of_week','is_weekend','lag_1','lag_2','lag_3','lag_6',
            'lag_12','lag_24','rolling_mean_3','rolling_std_3','rolling_mean_6',
            'rolling_std_6','rolling_mean_12','rolling_std_12','rolling_mean_24',
            'rolling_std_24','hour_sin','hour_cos','dow_sin','dow_cos','weather_score']

def predict_traffic(hour, day_of_week, temp, rain_1h, snow_1h, is_weekend,
                    clouds_all=0, day=1, month=1,
                    lag_1=0, lag_2=0, lag_3=0, lag_6=0, lag_12=0, lag_24=0,
                    rolling_mean_3=0, rolling_std_3=0,
                    rolling_mean_6=0, rolling_std_6=0,
                    rolling_mean_12=0, rolling_std_12=0,
                    rolling_mean_24=0, rolling_std_24=0,
                    hour_sin=0, hour_cos=0, dow_sin=0, dow_cos=0,
                    weather_score=0):
    values = [temp, rain_1h, snow_1h, clouds_all, hour, day, month,
              day_of_week, is_weekend, lag_1, lag_2, lag_3, lag_6,
              lag_12, lag_24, rolling_mean_3, rolling_std_3, rolling_mean_6,
              rolling_std_6, rolling_mean_12, rolling_std_12, rolling_mean_24,
              rolling_std_24, hour_sin, hour_cos, dow_sin, dow_cos, weather_score]

    X_new = pd.DataFrame([values], columns=features)

    lgb_pred = lgb_model.predict(X_new)[0]
    xgb_pred = xgb_model.predict(X_new)[0]
    final_pred = 0.55*lgb_pred + 0.45*xgb_pred

    return round(final_pred, 2)

iface = gr.Interface(
    fn=predict_traffic,
    inputs=[
        gr.Number(label="Hour (0-23)"),
        gr.Number(label="Day of Week (0=Mon,6=Sun)"),
        gr.Number(label="Temperature (Kelvin)"),
        gr.Number(label="Rain (mm in 1h)"),
        gr.Number(label="Snow (mm in 1h)"),
        gr.Checkbox(label="Is Weekend?")
    ],
    outputs="number",
    title="Traffic Volume Predictor",
    description="Enter weather and time features to predict traffic volume."
)

iface.launch()
```
